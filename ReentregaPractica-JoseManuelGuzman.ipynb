Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no compiencen por 28, ya que estamos con códigos postales de Madrid

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dnetro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

```r
airbnb$CodigoPostal[airbnb$CodigoPostal == ''] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '-'] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28'] <- NA
```

```r
airbnb$CodigoPostal[airbnb$CodigoPostal == '28002\n20882'] <- 28002
airbnb$CodigoPostal[airbnb$CodigoPostal == '2804'] <- 28004
airbnb$CodigoPostal[airbnb$CodigoPostal == '2805'] <- 28005
airbnb$CodigoPostal[airbnb$CodigoPostal == '2815'] <- 28015
```

```r
airbnb$CodigoPostal <- gsub('[^(28)]\\d{3}', NA, airbnb$CodigoPostal)
airbnb$CodigoPostal<-factor(airbnb$CodigoPostal)
str(airbnb$CodigoPostal)
```

Calcular un modelo lineal que combine alguna de estas variables:
* NumBanyos
* NumDormitorios
* MaxOcupantes
* MetrosCuadrados

```r
options(repr.plot.height=4,repr.plot.width=6)
barrio_sol <- subset(airbnb_entire, Barrio == "Sol")

model=lm(barrio_sol, formula=NumBanyos ~ NumDormitorios)


plot(barrio_sol$NumDormitorios,barrio_sol$NumBanyos, col="blue")
abline(c(0,model),col="red")
paste("El R^2 es: ",summary(model)$r.squared)
```

```r
options(repr.plot.height=4,repr.plot.width=6)
model=lm(barrio_sol, formula=NumDormitorios ~ MaxOcupantes)

plot(barrio_sol$MaxOcupantes,barrio_sol$NumDormitorios, col="blue")
abline(c(0,model),col="red")
paste("El R^2 es: ",summary(model)$r.squared)
#Buscad la correlación
#Limpiar los datos
#Los metros cuadrados no correlan muy bien porque hay pisos con 0 metros cuadrados < Filtrar algo que creamos que tiene sentido
# o POner NA
```

```r
options(repr.plot.height=4,repr.plot.width=6)
model=lm(barrio_sol, formula=NumDormitorios ~ MetrosCuadrados)

plot(barrio_sol$MetrosCuadrados,barrio_sol$NumDormitorios, col="blue")
abline(c(0,model),col="red")
paste("El R^2 es: ",summary(model)$r.squared)
```
